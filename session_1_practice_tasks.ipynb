## Section 9: Sets

### Task 9.1: Set Operations
Create two sets:
- `set_a = {1, 2, 3, 4, 5}`
- `set_b = {4, 5, 6, 7, 8}`

Calculate and print:
1. Union of both sets
2. Intersection of both sets
3. Difference (elements in set_a but not in set_b)
4. Symmetric difference (elements in either set but not both)


````python
# Write your code here
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

union = set_a | set_b

intersection = set_a & set_b

difference = set_a - set_b

symmetric_diff = set_a ^ set_b

print(f"Set A: {set_a}")
print(f"Set B: {set_b}")
print(f"Union: {union}")
print(f"Intersection: {intersection}")
print(f"Difference (A - B): {difference}")
print(f"Symmetric Difference: {symmetric_diff}")
````

### Challenge 1: Grade Calculator
Create a program that:
1. Asks the user for 5 test scores (use input and convert to integers)
2. Stores them in a list
3. Calculates and prints:
   - Average score
   - Highest score
   - Lowest score
   - Number of scores above 70


````python
# Write your code here

scores = []

print("Please enter 5 test scores:")
for i in range(5):
    score = __builtins__.int(input(f"Enter score {i+1}: "))
    scores.append(score)

average = sum(scores) / len(scores)
highest = max(scores)
lowest = min(scores)

above_70 = [s for s in scores if s > 70]
count_above_70 = len(above_70)

print("\n--- Results ---")
print(f"All Scores: {scores}")
print(f"Average Score: {average}")
print(f"Highest Score: {highest}")
print(f"Lowest Score: {lowest}")
print(f"Number of scores above 70: {count_above_70}")
````

### Challenge 2: Shopping Cart
Create a shopping cart system using a dictionary where:
- Keys are item names
- Values are dictionaries containing "price" and "quantity"

Example:
```python
cart = {
    "Apple": {"price": 2.5, "quantity": 3},
    "Bread": {"price": 3.0, "quantity": 1}
}
```

Calculate and print:
1. Total cost for each item (price Ã— quantity)
2. Total cost of the entire cart
3. Number of items in the cart


````python
# Write your code here

cart = {
    "Apple": {"price": 2.5, "quantity": 3},
    "Bread": {"price": 3.0, "quantity": 1},
    "Milk": {"price": 4.2, "quantity": 2},
    "Eggs": {"price": 0.5, "quantity": 12}
}

grand_total = 0
total_items_count = 0

print(f"{'Item':<10} | {'Qty':<5} | {'Price':<8} | {'Item Total'}")
print("-" * 45)

for item, details in cart.items():
    price = details["price"]
    quantity = details["quantity"]

    item_total = price * quantity

    grand_total += item_total
    total_items_count += quantity

    print(f"{item:<10} | {quantity:<5} | ${price:<7.2f} | ${item_total:>8.2f}")

print("-" * 45)

print(f"Total Unique Products: {len(cart)}")
print(f"Total Items in Cart:   {total_items_count}")
print(f"GRAND TOTAL:           ${grand_total:,.2f}")
````

### Challenge 3: Word Counter
Ask the user to enter a sentence. Then:
1. Split the sentence into words (use `.split()`)
2. Count how many words there are
3. Find the longest word
4. Create a dictionary showing how many times each word appears

**Hint:** Convert everything to lowercase for accurate counting


````python
# Write your code here

text = input("Enter a sentence: ")

words = text.lower().split()

word_count = len(words)

longest_word = max(words, key=len)

word_freq = {}
for word in words:

    word_freq[word] = word_freq.get(word, 0) + 1

print("\n--- Analysis Results ---")
print(f"Total words: {word_count}")
print(f"Longest word: '{longest_word}'")
print(f"Word Frequencies: {word_freq}")
````
